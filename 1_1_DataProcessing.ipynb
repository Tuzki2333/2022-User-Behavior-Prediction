```python
import pandas as pd
ad_feature = pd.read_csv('/content/drive/MyDrive/Taobao/ad_feature.csv')
raw_sample = pd.read_csv('/content/drive/MyDrive/Taobao/raw_sample.csv')
user_profile = pd.read_csv('/content/drive/MyDrive/Taobao/user_profile.csv')
# behavior_log = pd.read_csv('/content/drive/MyDrive/Taobao/behavior_log.csv')
```

```python
raw_sample = raw_sample.rename(columns={'user':'userid'})
```

```python
all_data = pd.merge(pd.merge(raw_sample,ad_feature),user_profile)
all_data['time'] = pd.to_datetime(all_data['time_stamp'], unit='s') + pd.Timedelta('08:00:00')
```

```python
all_data['day'] = all_data['time'].dt.day
all_data['hour'] = all_data['time'].dt.hour
```

```python
all_data.to_csv('/content/drive/MyDrive/Taobao/all_data.csv')
```
