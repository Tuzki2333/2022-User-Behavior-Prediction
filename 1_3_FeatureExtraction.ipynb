```
def get_results(original_df, group_idx_list, offset_idx, offset, new_column, aggregate_method = 'count'):
  if (aggregate_method == "count"):
    temp_df = original_df.groupby(group_idx_list)['clk'].count().reset_index()
  elif (aggregate_method == "sum"):
    temp_df = original_df.groupby(group_idx_list)['clk'].sum().reset_index()
  else:
    print('Error')
    return

  temp_df[offset_idx] += offset
  temp_df.rename(columns = {'clk':new_column}, inplace = True)

  return temp_df
```

```
# 过去一小时的展示次数
temp_num_show_last_hour = get_results(data, ['userid','adgroup_id','day_hour'], 'day_hour', 1, 'num_show_last_hour_user_adgroup','count')
# 过去一小时的点击次数
temp_num_click_last_hour = get_results(data, ['userid','adgroup_id','day_hour'], 'day_hour', 1, 'num_click_last_hour_user_adgroup','sum')
# 过去一天的展示次数
temp_num_show_last_day = get_results(data, ['userid','adgroup_id','day'], 'day', 1, 'num_show_last_day_user_adgroup','count')
# 过去一天的点击次数
temp_num_click_last_day = get_results(data, ['userid','adgroup_id','day'], 'day', 1, 'num_click_last_day_user_adgroup','sum')

data = pd.merge(data,temp_num_show_last_hour,how="left")
del temp_num_show_last_hour
gc.collect()

data = pd.merge(data,temp_num_click_last_hour,how="left")
del temp_num_click_last_hour
gc.collect()

data = pd.merge(data,temp_num_show_last_day,how="left")
del temp_num_show_last_day
gc.collect()

data = pd.merge(data,temp_num_click_last_day,how="left")
del temp_num_click_last_day
gc.collect()
```
