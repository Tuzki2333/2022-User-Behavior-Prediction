```
raw_sample = pd.read_csv('/content/drive/MyDrive/Taobao/raw_sample.csv')

raw_sample.describe()
```

```
raw_sample['user'] = raw_sample['user'].astype('category')

raw_sample[raw_sample['clk']==1]['user'].describe()
```

```
raw_sample['adgroup_id'] = raw_sample['adgroup_id'].astype('category')

raw_sample[raw_sample['clk']==1]['adgroup_id'].describe()
```

```
temp = pd.Categorical(raw_sample['clk'])
temp.categories = ['Not Clicked','Clicked']
temp.value_counts().plot.pie(autopct='%1.2f%%',fontsize = 18)
plt.ylabel('')
plt.tight_layout()
plt.savefig('/content/drive/MyDrive/Taobao/plot/ctr_pie.pdf',bbox_inches = 'tight')

# raw_sample['clk'].value_counts().plot.pie()
# plt.show()
```

```
raw_sample['time'] = pd.to_datetime(raw_sample['time_stamp'], unit='s') + pd.Timedelta('08:00:00')

raw_sample['day'] = raw_sample['time'].dt.day - 5
raw_sample['hour'] = raw_sample['time'].dt.hour
```

```
all_data['time'] = pd.to_datetime(all_data['time_stamp'], unit='s') + pd.Timedelta('08:00:00')

all_data['day'] = all_data['time'].dt.day - 5
all_data['hour'] = all_data['time'].dt.hour
```

```
plt.gca().xaxis.set_major_locator(MultipleLocator(4))
plt.plot(raw_sample.groupby('hour')['clk'].sum()/raw_sample.groupby('hour')['clk'].count(),'o-',label='Original Dataset')
plt.plot(all_data.groupby('hour')['clk'].sum()/all_data.groupby('hour')['clk'].count(),'o-', label='Dataset for Experiment')
plt.xlabel('Hour',fontsize = 18)
plt.ylabel('CTR',fontsize = 18)
plt.tight_layout()
plt.ylim(0.04,0.06)
plt.xticks(fontsize = 18)
plt.yticks(fontsize = 18)
plt.legend(fontsize = 16)
plt.savefig('/content/drive/MyDrive/Taobao/plot/ctr_hour.pdf',bbox_inches = 'tight')
```
